```python
import numpy as np
import copy
alpha=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']

def plugBoard(crypt):
    plug1=['a','b','c','d','e','f','g','h','i','j']
    plug2=['k','l','m','n','o','p','q','r','s','t']
    try:
        ind=plug1.index(crypt)
        return plug2[ind]
    except ValueError:
        try:
            ind=plug2.index(crypt)
            return plug1[ind]
        except ValueError:
            return crypt
    
def string_rotation(st, num):
    st=st[num:26]+st[:num]
    return st

def rotor1(crypt, setting, reverse):
    alpha=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    ind=alpha.index(crypt)
    #fixed connections=wiring
    order1=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    order2=['p','c','e','g','i','k','m','d','q','s','u','y','w','a','o','z','f','j','x','h','b','l','n','v','t','r']
    order1=string_rotation(order1, setting)
    order2=string_rotation(order2, setting)
    crypt=order1[ind]
    if(reverse==False):
        crypt=order2[ind]
    else:
        ind=order2.index(crypt)
        crypt=order1[ind]
    ind=order1.index(crypt)
    crypt=alpha[ind]
    return crypt
    
def rotor2(crypt, setting, reverse):
    alpha=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    ind=alpha.index(crypt)
    #fixed connections=wiring
    order1=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    order2=['f','b','k','e','l','t','j','s','v','y','c','m','i','x','u','n','d','r','h','a','o','q','z','g','w','p']
    order1=string_rotation(order1, setting)
    order2=string_rotation(order2, setting)
    crypt=order1[ind]
    if(reverse==False):
        crypt=order2[ind]
    else:
        ind=order2.index(crypt)
        crypt=order1[ind]
    ind=order1.index(crypt)
    crypt=alpha[ind]
    return crypt
    
def rotor3(crypt, setting, reverse):
    alpha=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    ind=alpha.index(crypt)
    #fixed connections=wiring
    order1=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    order2=['k','f','l','n','g','m','h','e','r','w','a','o','u','p','x','z','i','y','v','t','q','b','j','c','s','d']
    order1=string_rotation(order1, setting)
    order2=string_rotation(order2, setting)
    crypt=order1[ind]
    if(reverse==False):
        crypt=order2[ind]
    else:
        ind=order2.index(crypt)
        crypt=order1[ind]
    ind=order1.index(crypt)
    crypt=alpha[ind]
    return crypt

    
def loopBack(crypt):
    if(crypt=='a'):
        crypt='y'
    elif(crypt=='y'):
        crypt='a'
    elif(crypt=='b'):
        crypt='r'
    elif(crypt=='r'):
        crypt='b'
    elif(crypt=='c'):
        crypt='u'
    elif(crypt=='u'):
        crypt='c'
    elif(crypt=='d'):
        crypt='h'
    elif(crypt=='h'):
        crypt='d'
    elif(crypt=='e'):
        crypt='q'
    elif(crypt=='q'):
        crypt='e'
    elif(crypt=='f'):
        crypt='s'
    elif(crypt=='s'):
        crypt='f'
    elif(crypt=='g'):
        crypt='l'
    elif(crypt=='l'):
        crypt='g'
    elif(crypt=='i'):
        crypt='p'
    elif(crypt=='p'):
        crypt='i'
    elif(crypt=='j'):
        crypt='x'
    elif(crypt=='x'):
        crypt='j'
    elif(crypt=='k'):
        crypt='n'
    elif(crypt=='n'):
        crypt='k'
    elif(crypt=='m'):
        crypt='o'
    elif(crypt=='o'):
        crypt='m'
    elif(crypt=='t'):
        crypt='z'
    elif(crypt=='z'):
        crypt='t'
    elif(crypt=='v'):
        crypt='w'
    elif(crypt=='w'):
        crypt='v'
    else:
        print('something went wrong pretty bad')
    return crypt
def setting_updation(settings):
    if(settings[2]==25):
        settings[2]=0
        if(settings[1]==25):
            settings[1]=0
            if(settings[0]==25):
                settings[0]=0
            else:
                settings[0]+=1
        else:
            settings[1]+=1
    else:
        settings[2]+=1
def rotors_box(crypt, settings):
    crypt=rotor1(crypt, settings[2], False)
    crypt=rotor2(crypt, settings[1], False)
    crypt=rotor3(crypt, settings[0], False)
    crypt=loopBack(crypt)
    crypt=rotor3(crypt, settings[0], True)
    crypt=rotor2(crypt, settings[1], True)
    crypt=rotor1(crypt, settings[2], True)
    setting_updation(settings)
    return crypt
def encryption(st, settings):
    encrypted=''
    for ch in st:
        crypt=ch
        crypt=plugBoard(crypt)
        crypt=rotors_box(crypt, settings)
        crypt=plugBoard(crypt)
        encrypted+=crypt
    return encrypted
def decryption(st, settings):
    decrypted=''
    for ch in st:
        crypt=ch
        crypt=plugBoard(crypt)
        crypt=rotors_box(crypt, settings)
        crypt=plugBoard(crypt)
        decrypted+=crypt
    return decrypted
```

```python
to_be_crypted_string='malayalam'
settings=[13, 5, 3]
encrypted=encryption(to_be_crypted_string, settings)
print('encrypted=', encrypted)
settings=[13, 5, 3]
decrypted=decryption(encrypted, settings)
print('decrypted=', decrypted)
```
